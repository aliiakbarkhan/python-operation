```python
import random

coin = random.choice(["ali", "hasan"])
print(coin)
```

```python
from random import choice as cs

coin = cs(["ali","hasan","aslam"])
print(coin)
```

```python
# This function will generate a random number between 1 and 19
from random import randint as rn
def genert():
  number = rn(1,19)
  print(number)
genert()
```

```python
# it will shuffle a list into a random order.
from random import shuffle as sh
card = ["ali", "hasan", "aslam"]
sh(card)
for _ in card:
  print(_)
```

```python
from statistics import mean as m
number = m([100,90])
print(number)
```
