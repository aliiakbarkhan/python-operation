```python
items = {
    "Baja Taco": 4.25,
    "Burrito": 7.50,
    "Bowl": 8.50,
    "Nachos": 11.00,
    "Quesadilla": 8.50,
    "Super Burrito": 8.50,
    "Super Quesadilla": 9.50,
    "Taco": 3.00,
    "Tortilla Salad": 8.00
}
def func():
  x2 = 0
  while True:
    try:
      x = input("Item: ")
      x = x.title()
    except KeyError:
      func()

    else:
      if x not in items:
        func()
      else:
       x2 += items[x]
       print("Total:",f"${x2}")

func()
```
