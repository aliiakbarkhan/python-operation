```python
while True:
  try:
    n,d = (input("Fraction")).split("/")
    if d!=0:
      ans = (int(n)/int(d))*100
      if ans == 100:
        print("F")
      elif ans == 0:
        print("E")
      elif ans>0 and ans<100:
        print(ans)
    else:
      continue


  except (ValueError, ZeroDivisionError):
    pass
  else:
    break
```
