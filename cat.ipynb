```python
i = 4
while i != 0:
  print("meow")
  i = i-1
```

```python
i = 1
while i <= 4:
  print("meow")
  i = i+1
```

```python
# i += 1 is the same as saying i = i + 1
i = 0
while i < 3:
  print("meow")
  i+=1
```
